```python
%%writefile app.py
import streamlit as st
import tempfile
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain_chroma import Chroma
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Get API key from .env file
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

st.title("CHATBOT")

# File uploader
upload_file = st.file_uploader("Upload the PDF file", type=["pdf"])

if upload_file:
    st.success("✅ PDF received successfully")

    # Save uploaded file to a temporary path
    with tempfile.NamedTemporaryFile(delete=False, suffix=".pdf") as tmp_file:
        tmp_file.write(upload_file.read())
        tmp_path = tmp_file.name

    # Load PDF and split into chunks
    loader = PyPDFLoader(tmp_path)
    pages = loader.load()
    splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=15)
    docs = splitter.split_documents(pages)

    # Create embeddings and vector database
    embeddings = OpenAIEmbeddings(api_key=OPENAI_API_KEY)
    vector_db = Chroma.from_documents(documents=docs, embedding=embeddings, persist_directory="new_dbstory")
    retriever = vector_db.as_retriever()

    # Memory for conversation
    memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True, output_key="answer")

    # LLM
    llm = ChatOpenAI(api_key=OPENAI_API_KEY, model="gpt-4o-mini")

    # Conversational Retrieval Chain
    chat_chain = ConversationalRetrievalChain.from_llm(
        llm=llm,
        retriever=retriever,
        memory=memory,
        return_source_documents=True
    )

    # Input box for user query
    user_input = st.text_input("Ask a question about the PDF:")

    if user_input:
        response = chat_chain.invoke({"question": user_input})
        st.subheader("🤖 Answer:")
        st.write(response["answer"])

#         with st.expander("📄 Source documents"):
#             for i, doc in enumerate(response["source_documents"]):
#                 st.markdown(f"**Page {i+1}**")
#                 st.write(doc.page_content)

# else:
#     st.info("📥 Please upload a PDF file to start.")
```
