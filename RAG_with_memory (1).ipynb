```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_openai import ChatOpenAI
llm=ChatOpenAI(api_key=Open_API_Key,temperature=0)
```

```python
loder=PyPDFLoader("/content/DS CoverLetter.pdf")
doc=loder.load()
doc
```

```python
spliter=RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)
chunk=spliter.split_documents(doc)
chunk
```

```python
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain_chroma import Chroma
embeddings=OpenAIEmbeddings(api_key=Open_API_Key)
embeddings
```

```python
# save to chroma
vector_Db=Chroma.from_documents(
    documents=chunk,
    embedding=embeddings,
    persist_directory="conrag_pdf")
vector_Db.add_documents(chunk)
```

```python
retriever = vector_Db.as_retriever()
```

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True, output_key="answer")
```

```python
from langchain.chains import ConversationalRetrievalChain

chat_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory,
    return_source_documents=True
)
# plain methodee
# response1 = chat_chain.invoke({"question":"Title of the document and document represent who carreier?"})
# print(response1["answer"])

# # Second question – remembers context from the first!
# response2 = chat_chain.invoke("What is his background?")
# print(response2["answer"])

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Exiting chat.")
        break

    response = chat_chain.invoke({"question": user_input})
    print("AI:", response["answer"])
```
